```python
import time

class TuringMachine(object):
    
    def __init__(self):
        self.states = []
        self.alphabet = []
        self.symbol_tape = []
        self.final_states = -1
        self.transitions = dict()
        self.test_cases = []

    def __call__(self, filename):
        
        file = open(filename, 'r')
        lines = file.readlines()
        num_states = int(lines[0])
        
        if num_states > 0 and num_states <= 10:
            self.states = [str(s) for s in range(0, num_states)]
        else:
             print("Error")
            
        #take the information about the number of simbols and the simbols of the alphabet
        info_alphabet = lines[1].strip().split(' ')
        num_alphabet = int(info_alphabet[0])
        
        if num_alphabet > 0 and num_alphabet <= 10:
            self.alphabet = info_alphabet[1:num_alphabet+1] + ["-","B"]
        else:
             print("Error")
            
            
        info_symbol_tape = lines[2].strip().split(' ')
        
        num_symbol_tape = int(info_symbol_tape[0])
        
        if num_symbol_tape > 0 and num_symbol_tape <= 10:
            self.symbol_tape = info_symbol_tape[1:num_symbol_tape]
            
        self.final_states = int(lines[3]) 
        
        num_transitions = int(lines[4])
        
        # 0 a Z 0 AZ
        if num_transitions > 0 and num_transitions <=50:
            for i in range(num_transitions):
                    info_transition = lines[5 + i].split()
                    q = int(info_transition[0]) #current state
                    x = info_transition[1] # alphabet simbol
                    p = int(info_transition[2]) # next state
                    Y = info_transition[3] # simbol tape
                    D = info_transition[4] # direction of transition
                    #print p, x, Z, "=", q, o
                    self.add_transition(q, x, p, Y, D)
                
                
        num_test_cases = int(lines[5+num_transitions])
        
        if num_test_cases > 0 and num_test_cases <= 40:
            
            for i in range(1, num_test_cases+1):
                #point to the first case for the string to test
                test_case = lines[5+num_transitions+i].strip()
                self.test_cases.append(test_case)
            
        

    def add_transition(self, q, x, p, Y, D):
        
        
        if self.transitions.get(q, -1) == -1:
            self.transitions[q] = {}
        if self.transitions[q].get(x, -1) == -1:
            self.transitions[q][x] = {}
       
        self.transitions[q][x] = (p,Y,D)

    def evaluate(self, string):
        tic = time.time()
        current_state = int(self.states[0])
        string = "B" + string + "B"
        pointer = 1
        final_state = int(self.final_states)
        while current_state != final_state:
            if 0 <= pointer < len(string):
                c = str(string[pointer])
                if self.transitions.get(current_state, -1) != -1 and self.transitions[current_state].get(c,-1) != -1:
                    print(current_state, c, "-->", self.transitions[current_state][c])
                    (p, Y, D) = self.transitions[current_state][c]
                    current_state = p
                    
                    string = string[:pointer]+Y+string[pointer+1:]
                    
                    print(string)
                       
                    if D == "L":
                        pointer = pointer-1
                    elif D == "R":
                        pointer = pointer+1
                else:
                    print(current_state, c, "-->","Transition not found.")
                    break
            else:
                print("Undefined state")
                break
        toc = time.time()
        print("Time: %f seconds" % (toc - tic))
        
        return "aceita" if bool(current_state == final_state) else "rejeita"


if __name__ == '__main__':
    turing_machine = TuringMachine()
    
    turing_machine('tape_tm.txt')
    
    for test_case in turing_machine.test_cases:
        print (test_case, turing_machine.evaluate(test_case))
```
